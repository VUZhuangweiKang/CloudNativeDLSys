```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
title = open('./logs/log1.txt').readline()
data = pd.read_csv('./logs/log1.csv')
x = np.arange(len(np.unique(data['file_size'])))
for total_size, group in data.groupby(by='total_size'):
    group.sort_values(by='file_size', inplace=True)
    plt.plot(x, group['total_time'], label=f"total_size={total_size}GB", marker='*')

plt.xticks(x, np.unique(data['file_size']), rotation=30)
plt.xlabel('block size (KB)')
plt.ylabel('load time(s)')
plt.title(title)
plt.legend()
plt.grid()
plt.show()
```

```python
data = pd.read_csv('./logs/log7.csv')
x = np.arange(len(np.unique(data['file_size'])))
for total_size, group1 in data.groupby('total_size'):
    for bandwidth, group2 in group1.groupby('bandwidth'):
        group2.sort_values(by='file_size', inplace=True)
        plt.plot(x, group2['total_time'], label=f"total_size={total_size}GB, bandwidth={bandwidth}Mbps")

    plt.xticks(x, np.unique(data['file_size']), rotation=30)
    plt.xlabel('block size (KB)')
    plt.ylabel('load time(s)')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
def plot(file):
    data = pd.read_csv(file)
    data['block_size'] /= 1e6
    # data['latency'] *= 1000
    data.sort_values(by='read_rate', inplace=True, ascending=False)
    for read_rate, group in data.groupby(by="read_rate"):
        group.sort_values(by='block_size', inplace=True)
        plt.scatter(group['block_size'], group['latency'])
    return group
```
