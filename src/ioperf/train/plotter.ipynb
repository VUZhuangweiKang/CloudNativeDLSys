```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = pd.read_csv('ioperf.csv', header=None, delimiter=' ', names=['bandwidth', 'chunkSize', 'iomode', 'latency'])
local_io = data[data['iomode']=='local']
remote_io = data[data['iomode']=='remote']
local_io.sort_values(by=['chunkSize'], inplace=True)
remote_io.sort_values(by=['chunkSize'], inplace=True)
```

```python
local_io.shape, remote_io.shape
```

```python
fig = plt.figure(figsize=(15, 5))
plt.scatter(local_io['chunkSize'], local_io['latency'])
plt.show()
```

```python
fig = plt.figure(figsize=(15, 5))
plt.scatter(remote_io['chunkSize'], remote_io['latency'])
plt.show()
```
